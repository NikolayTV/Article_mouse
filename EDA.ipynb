```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('df2.csv', header=[0,1])
# df = df.drop('проба', axis=1).set_index(['№', 'группа', 'шифр крысы'])

df = df.drop('проба', axis=1)
df = df.set_index([('№', '0'), ('группа', '0'), ('шифр крысы', '0')])
print(df.shape)
df

# Все что меньше 4.9 должно быть 4.9
```

```python
ref_df = pd.read_csv('reference_values.csv', index_col=0)
ref_df
```

```python
# Binarize data 
for name in ref_df.index:
    for day_num in ['0', '3', '7', '14']:
        df.loc[df.index, (name, day_num)] = pd.cut(x = df[name][day_num], 
                                             bins=ref_df.loc[name].values, 
                                             labels=[str(i) for i in range(1,8)]).values
df        
```

```python
# Количество пропусков по строкам
df.isna().sum(axis=1)
```

```python
print('Количество пропусков в столбцах')
df.isna().sum(axis=0).value_counts().sort_index()
```

```python
# Исключаем строки в которых больше 25 пропусков
df = df[df.isna().sum(axis=1) < 25]
df.shape
```

```python
# Заполнить пропуски самым популярным значением категории
for col in set(df.columns.get_level_values(0)):
    for day_num in ['0', '3', '7', '14']:
        df[col][day_num].value_counts()
        most_popular = df[col][day_num].value_counts().index[0]
        df.loc[df.index, (col, day_num)] = df[col][day_num].fillna(most_popular)
```

```python
df.to_pickle('df_no_nans.pkl')
```

```python
# Посчитать количество уникальных значений в столбцах
for col in df.columns:
    df[col] = df[col].astype(int)

unique_list = []
for col in set(df.columns.get_level_values(0)):
    for day_num in ['0', '3', '7', '14']:
        unique_num = df[col][day_num].value_counts().shape[0]
        unique_list.append(unique_num)

plt.hist(unique_list)
```

# Spearman correlation

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# sns.set_theme(style="white")

# Compute the correlation matrix
corr = df.corr(method='spearman')

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)



# Draw the heatmap with the mask and correct aspect ratio
plt.figure(figsize=(15,15));
sns.heatmap(corr, mask=mask, cmap=cmap,  center=0, #vmax=.3,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

```python
from scipy import stats

mat = df.values.T
K = len(df.columns)
correl = np.empty((K,K), dtype=float)
p_vals = np.empty((K,K), dtype=float)

for i, ac in enumerate(mat):
    for j, bc in enumerate(mat):
        if i > j:
            continue
        else:
            corr = stats.spearmanr(ac, bc)
            #corr = stats.kendalltau(ac, bc)

        correl[i,j] = corr[0]
        correl[j,i] = corr[0]
        p_vals[i,j] = corr[1]
        p_vals[j,i] = corr[1]

df_p = pd.DataFrame(p_vals, columns=df.columns, index=df.columns)
df_corr = pd.DataFrame(correl, columns=df.columns, index=df.columns)
```

```python
def correlation(corr_matrix, threshold):
    col_corr = set() # Set of all the names of deleted columns
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] <= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
    return col_corr

col_corr = correlation(df_p, 0.05)

# col_corr
df_corr_reduced = df_corr[df_p <= 0.05]
df_corr_reduced
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(df_corr_reduced, dtype=bool))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
plt.figure(figsize=(15,15));
sns.heatmap(df_corr_reduced, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

# Kendall correlation

```python
# Compute the correlation matrix
corr = df.corr(method='kendall')

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
plt.figure(figsize=(15,15));
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
mat = df.values.T
K = len(df.columns)
correl = np.empty((K,K), dtype=float)
p_vals = np.empty((K,K), dtype=float)

for i, ac in enumerate(mat):
    for j, bc in enumerate(mat):
        if i > j:
            continue
        else:
            corr = stats.kendalltau(ac, bc)

        correl[i,j] = corr[0]
        correl[j,i] = corr[0]
        p_vals[i,j] = corr[1]
        p_vals[j,i] = corr[1]

df_p = pd.DataFrame(p_vals, columns=df.columns, index=df.columns)
df_corr = pd.DataFrame(correl, columns=df.columns, index=df.columns)
```

```python
col_corr = correlation(df_p, 0.05)
df_corr_reduced = df_corr[df_p < 0.05]
df_corr_reduced
```

# Корреляция фридмана о динамике по дням

```python
from scipy.stats import friedmanchisquare

df = pd.read_pickle('df_no_nans.pkl')
friedman_df = pd.DataFrame(index = df.columns.get_level_values(0), columns = ['friedman_stat', 'p'])

for col in set(df.columns.get_level_values(0)):
    data1 = df[col]['0'].values       
    data2 = df[col]['3'].values       
    data3 = df[col]['7'].values       
    data4 = df[col]['14'].values       

    stat, p = friedmanchisquare(data1, data2, data3, data4)
    friedman_df.loc[col, ['friedman_stat', 'p']] = stat, p


friedman_df
```
